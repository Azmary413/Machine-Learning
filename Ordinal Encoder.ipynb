```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("tree2.csv")
```

```python
df
```

```python
df["Income"].unique()
```

```python
Income_order = ["Low", "Medium", "High"]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
enco = OrdinalEncoder(categories = [Income_order])
```

```python
enco.fit(df[["Income"]])
```

```python
final = pd.DataFrame(enco.transform(df[["Income"]]))
```

```python
df["Income"] = final
```
