```python
import pandas as pd
```

```python
df = pd.read_csv("car.csv")
```

```python
df
```

```python
dum = pd.get_dummies(df["Car Model"])
```

```python
dum
```

```python
merge = pd.concat([df,dum], axis = "columns")
```

```python
merge
```

```python
final = merge.drop(["Car Model"], axis = "columns")
```

```python
final
```

```python
y = df["Sell Price"]
```

```python
y
```

```python
x = final.drop(["Sell Price"], axis = 1)
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[87650, 5, 0, 1, 0, 0]])
```
