```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("car.csv")
```

```python
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

# Seperate x, y

```python
x = df.drop(["Sell Price"], axis = 1)
```

```python
x
```

```python
y = df["Sell Price"]
```

```python
y
```

# One-Hot Encoding

```python
dum = pd.get_dummies(x["Car Model"])
```

```python
dum
```

```python
merge = pd.concat([x,dum], axis = 1)
```

```python
merge
```

```python
final = merge.drop(["Car Model","Age"], axis = "columns")
```

```python
final
```

# Split the dataset for training and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
finaltrain, finaltest, ytrain, ytest = train_test_split(final,y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
obj = LinearRegression()
```

```python
obj.fit(finaltrain, ytrain)
```

```python
finaltest
```

```python
ytest
```

```python
#Predict the value
pre = obj.predict(finaltest)
```

```python
pre
```

```python
obj.score(finaltest,ytest)
```

# R-Squared Value

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(ytest,pre)
```
